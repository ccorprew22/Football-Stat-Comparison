```python
style.use('dark_background')

Todd_stats = pd.read_csv('ToddGurley_Stat.csv')
Todd_stats.set_index("DATE", inplace=True)
Todd_stats.drop("REGULAR SEASON STATS", axis=0, inplace=True)
Todd_stats= Todd_stats[::-1] #reverse chart so it goes in date order
td_total1 = []
total1 = 0
#Adding touchdowns so that it grows each week
for td in Todd_stats['TD']:
    total1 += td
    td_total1.append(total1)
Todd_stats['TG_Total TD'] = td_total1

LT_stats = pd.read_csv('LT_Stat.csv')
LT_stats.set_index("DATE", inplace = True)
LT_stats = LT_stats[::-1]
td_total2 = []
total2 = 0
for td in LT_stats['TD']:
    total2 += td
    td_total2.append(total2)
LT_stats['LT_Total TD'] = td_total2

ax = Todd_stats['TG_Total TD'].plot()
#PRESENTS THE SAME NUMBER OF WEEKS
LT_stats.head(11)['LT_Total TD'].plot(ax = ax)
plt.legend()
plt.show()

print("TG \n")
print(Todd_stats)
print('\n')
print("LT \n")
print(LT_stats.head(11))
```
